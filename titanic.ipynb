```python
import pandas as pd
import numpy as np
```

```python
df_train=pd.read_csv('train.csv')
```

```python
df_train.head
```

```python
y_train=df_train[['Survived']]
```

```python
x_train=df_train.drop(['PassengerId','Survived','Name'],axis=1)
```

```python

x_train=x_train.drop(['Cabin','Ticket'],axis=1)
```

```python
x_train.info()
x_train.head(10)
```

```python
x_train['Sex']=x_train['Sex'].map({'male':0,'female':1})
```

```python
x_train['Embarked']=x_train['Embarked'].map({'S':0,'Q':1,'C':2})
```

```python
x_train['Age']=x_train['Age'].fillna(0)
x_train['Fare']=x_train['Fare'].fillna(0)
x_train['Embarked']=x_train['Embarked'].fillna(0)
```

```python
import sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
clf=RandomForestClassifier(n_jobs=-1,random_state=42)

clf.fit(x_train,df_train['Survived'])
```

```python
df_test=pd.read_csv('test.csv')
```

```python
df_test.head(10)
```

```python
df_test=df_test.drop(['Cabin','Ticket','Name'],axis=1)
```

```python
df_test=df_test.drop(['PassengerId'],axis=1)
df_test.head(10)
```

```python
df_test['Sex']=df_test['Sex'].map({'male':0,'female':1})
df_test['Embarked']=df_test['Embarked'].map({'S':0,'Q':1,'C':2})
```

```python
df_test['Age']=df_test['Age'].fillna(0)
df_test['Fare']=df_test['Fare'].fillna(0)
df_test['Embarked']=df_test['Embarked'].fillna(0)
```

```python
y=clf.predict(df_test)
```

```python
new_test=pd.read_csv('test.csv')
Pid=new_test[['PassengerId']]
```

```python
final_answer=pd.DataFrame(Pid)
```

```python
final_answer['Survived']=y
final_answer
```
